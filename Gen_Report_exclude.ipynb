```python
# EEG Data Quality Control Analysis - Complete Pipeline
# Based on the Phase 1-3 Quality Control Checklist
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os
import sys
import argparse
import glob
import cmlreaders as cml
import json
from matplotlib.ticker import FuncFormatter
import warnings
import re
from scipy import signal
from scipy.stats import zscore

# Configure pandas and warnings
pd.set_option('display.max_columns', 1000)
warnings.filterwarnings("ignore")
warnings.simplefilter('ignore')
df = cml.get_data_index()
# Change directory location if needed:    


data_dir = "/scratch/new_courier_pilot/"  # pilot data directory
# data_dir = "/scratch/EFRCourier_reports/"  # hospital data directory
time_window = 10000  # how long to visualize the data
exp = 'EFRCourierOpenLoop'
df_select=df[df['experiment'] == exp]
allsubject=df_select['subject'].unique()
print(f'{allsubject}')
```

```python
subject ='R1620J'
df_select_new = df_select[(df_select['subject'] == subject)]
df_select_new['subject'].unique()
df_select_new['session'].nunique()
```

```python
# =============================================================================
# CONFIGURATION
# =============================================================================
for i in range(df_select_new['session'].nunique()):
    print("="*80)
    print(f"EEG DATA Report for subject:{subject}")
    print(f"Experiment: {exp}")
    print(f"Time window: {time_window} ms")
    print(f"Processing session {i+1} of {len(df_select_new)}")
    print("-"*80)
    
    # Get the current session data 
    df_sess = df_select_new.iloc[i]
    reader = cml.CMLReader(subject, exp, session=df_sess['session'],
                          localization=df_sess['localization'], 
                          montage=df_sess['montage'])
    
    # =============================================================================
    # PHASE 1: DATA LOGIN VERIFICATION
    # =============================================================================
    print("\nPHASE 1: DATA LOGIN VERIFICATION")
    print("-"*50)

    # Load data index and get subject data

    if df_select.empty:
        raise ValueError(f"No data found for subject {subject} in experiment {exp}")

    # Initialize reader and load events
    evs = reader.load('task_events')

    print("1.1 EVENT VERIFICATION")
    print("-"*30)

    # Define expected event counts  
    expected_counts = {
        'START': 1,
        'store mappings': 1,
        'SESS_END': 1,
        'TRIAL_START': 5,
        'TRIAL_END': 5,
        'WORD': 60,
        'STIM': '89 or 91',
        'pointing begins': 65,
        'pointing finished': 65,
        'REC_START': 5,
        'REC_STOP': 5,
        'CUED_REC_CUE': 60,
        'CUED_REC_STOP': 60
    }

    # Verify event counts
    event_verification = {}
    for event_type, expected in expected_counts.items():
        actual = len(evs[evs['type'] == event_type])
        event_verification[event_type] = {
            'actual': actual,
            'expected': expected,
            'status': 'PASS' if (isinstance(expected, int) and actual == expected) or 
                               (isinstance(expected, str) and actual in [89, 91]) else 'FAIL'
        }
        print(f"{event_type}: {actual} / {expected} - {event_verification[event_type]['status']}")

    print("\n1.2 STIMULUS DELIVERY VALIDATION")
    print("-"*30)

    # Extract stimulus parameters
    pulse_freq_data = []
    word_data_type = []
    for i in range(len(evs['stim_params'])):
        freq_data = evs['stim_params'].iloc[i]['burst_freq']
        word_data = evs['type'].iloc[i]
        pulse_freq_data.append(freq_data)
        word_data_type.append(word_data)

    # Find stimulus indices
    indices_with_8 = np.where(np.array(pulse_freq_data) == 8)[0]
    indices_with_8=indices_with_8.tolist() 
    indices_with_3 = np.where(np.array(pulse_freq_data) == 3)[0]
    indices_with_3=indices_with_3.tolist() 
    indices_with_no = np.where(np.array(pulse_freq_data)[:-1] == -1)[0] + 1
    indices_with_no = indices_with_no.tolist()

    print(f"8 Hz stimuli events: {len(indices_with_8)}")
    print(f"3 Hz stimuli events: {len(indices_with_3)}")
    print(f"No stimuli events: {len(indices_with_no)}")

    # Add phase information to events
   # def add_phase(evs):
    def add_phase(evs):
        new_evs = pd.DataFrame([])
        for ind,list_evs in evs.groupby('trial'): # 'trial' is the aquivalent of a delivery day
            if list_evs['trial'].unique()!=-999:
                list_evs['phase']='nan'
            if (not list_evs[list_evs['type'] == 'TRIAL_START'].index.empty and 
            not list_evs[list_evs['type'] == 'TRIAL_END'].index.empty):
                # find indexes of encoding start and end
                first_pointing_index = min(list_evs[list_evs['type']=='TRIAL_START'].index)
                navigation_end_index = min(list_evs[list_evs['type']=='TRIAL_END'].index)-1
                list_evs.loc[first_pointing_index:navigation_end_index]['phase']='encoding'
                # find indexes of retrieval start and end
                rec_start_index = list_evs[list_evs['type']=='REC_START'].index[0]
                rec_end_index = list_evs[list_evs['type']=='REC_STOP'].index[0]
                list_evs.loc[rec_start_index:rec_end_index]['phase']='retrieval'

                if len(list_evs[list_evs.type=='FSR_START'])>0: # relevant only for the final trial
                    # find indexes of Final FR for stores start and end
                    FSR_start_index = list_evs[list_evs['type']=='FSR_START'].index[0]
                    FSR_end_index = list_evs[list_evs['type']=='FSR_STOP'].index[0]
                    list_evs.loc[FSR_start_index:FSR_end_index]['phase']='FSR'

                     # find indexes of Final FR for items start and end
                    FFR_start_index = list_evs[list_evs['type']=='FFR_START'].index[0]
                    FFR_end_index = list_evs[list_evs['type']=='FFR_STOP'].index[0]
                    list_evs.loc[FFR_start_index:FFR_end_index]['phase']='FFR'

            elif list_evs['trial'].unique()==-999:
                list_evs['phase']='nan'


            new_evs=pd.concat([new_evs,list_evs])
        return new_evs
    evs=add_phase(evs)
    # Analyze stimuli by phase
    result_8HZ_encoding = evs[(evs.index.isin(indices_with_8)) & (evs['phase'] == 'encoding')]
    result_3HZ_encoding = evs[(evs.index.isin(indices_with_3)) & (evs['phase'] == 'encoding')] 
    result_none_encoding = evs[(evs.index.isin(indices_with_no)) & (evs['phase'] == 'encoding') & (evs['type'] == 'WORD')]
    result_8HZ_retrieval = evs[(evs.index.isin(indices_with_8)) & (evs['phase'] == 'retrieval')]
    result_3HZ_retrieval = evs[(evs.index.isin(indices_with_3)) & (evs['phase'] == 'retrieval')]
    print(f"\nStimuli distribution by phase:")
    print(f"Encoding - 3 Hz: {len(result_3HZ_encoding)}, 8 Hz: {len(result_8HZ_encoding)}, None: {len(result_none_encoding)}")
    print(f"Retrieval - 3 Hz: {len(result_3HZ_retrieval)}, 8 Hz: {len(result_8HZ_retrieval)}")

    # =============================================================================
    # PHASE 2: FREQUENCY ANALYSIS (3-8 Hz Peak Detection)
    # =============================================================================
    print("\n" + "="*80)
    print("PHASE 2: FREQUENCY ANALYSIS (3-8 Hz Peak Detection)")
    print("-"*50)

    # Load EEG data - reader is already initialized
    evs = reader.load("events")
    channels = reader.load("pairs")

    # Load EEG for different conditions
    print("Loading EEG data...")
    eeg_container_8HZ_encoding = reader.load_eeg(result_8HZ_encoding, -1000, time_window, scheme=channels)
    eeg_container_3HZ_encoding = reader.load_eeg(result_3HZ_encoding, -1000, time_window, scheme=channels)
    eeg_container_8HZ_retrieval = reader.load_eeg(result_8HZ_retrieval, -1000, time_window, scheme=channels)
    eeg_container_3HZ_retrieval = reader.load_eeg(result_3HZ_retrieval, -1000, time_window, scheme=channels)
    eeg_container_no_encoding = reader.load_eeg(result_none_encoding, -1000, time_window, scheme=channels)

    # Extract data arrays
    eeg_8HZ = eeg_container_8HZ_encoding.data
    eeg_3HZ = eeg_container_3HZ_encoding.data
    eeg_8HZ_retrieval = eeg_container_8HZ_retrieval.data
    eeg_3HZ_retrieval = eeg_container_3HZ_retrieval.data
    eeg_no_encoding = eeg_container_no_encoding.data
    sr = eeg_container_8HZ_encoding.samplerate

    print(f"Sample rate: {sr} Hz")
    print(f"EEG data shapes:")
    print(f"  8 Hz encoding: {eeg_8HZ.shape}")
    print(f"  3 Hz encoding: {eeg_3HZ.shape}")
    print(f"  No stimuli encoding: {eeg_no_encoding.shape}")

    # Identify stimulated channels
    def identify_stimulated_channels(evs, indices, channels):
        """Identify stimulated channel pairs and related channels"""
        labels = []
        Stim_label=[]
        for ind in indices:
            stim_params = evs['stim_params'].loc[ind]
            anode_label = stim_params['anode_label']
            cathode_label = stim_params['cathode_label']
            label1 = anode_label + '-' + cathode_label
            label2 = cathode_label + '-' + anode_label
            labels.extend([label1, label2])
            Stim_label.extend([anode_label,cathode_label])
        labels = list(set(labels))
        Stim_label = list(set(Stim_label))# Remove duplicates

        all_stimuli_channel_indices = []
        all_stimuli_indices = []
        relative_indices = []
        for label in labels:
            # Exact match
            mask = channels['label'].str.contains(label, na=False)
            stimuli_indices = channels.loc[mask].index.tolist()
          
         
            # Flexible match - remove numbers from both search term and data
            cleaned_label = re.sub(r'\d+', '', label)
            mask_channel = channels['label'].str.replace(r'\d+', '', regex=True).str.contains(cleaned_label, na=False)
            stimuli_channel_indices = channels.loc[mask_channel].index.tolist()

            all_stimuli_indices.extend(stimuli_indices)
            all_stimuli_channel_indices.extend(stimuli_channel_indices)
        for label in Stim_label:
            # Exact match
            mask = channels['label'].str.contains(label, na=False)
            relative_index = channels.loc[mask].index.tolist()
            relative_indices.extend(relative_index)
           
        
        all_stimuli_indices = list(set(all_stimuli_indices))
        all_stimuli_channel_indices = list(set(all_stimuli_channel_indices))
        relative_index = list(set(relative_index))
        ALLother = list(range(1, len(channels)))
        Index_other = [x for x in ALLother if x not in all_stimuli_channel_indices]
        Contact_stimuli = stimuli_indices
        Channel_stimuli = [x for x in all_stimuli_channel_indices 
                  if x not in all_stimuli_indices 
                  and x not in relative_indices]
        return Index_other, Channel_stimuli, Contact_stimuli

    # Identify channels for both frequencies
    Index_other_8, Channel_stimuli_8, Contact_stimuli_8 = identify_stimulated_channels(evs, indices_with_8, channels)
    Index_other_3, Channel_stimuli_3, Contact_stimuli_3 = identify_stimulated_channels(evs, indices_with_3, channels)

    print(f"\n8 Hz stimulated contacts: {Contact_stimuli_8}")
    print(f"3 Hz stimulated contacts: {Contact_stimuli_3}")

    # =============================================================================
    # VISUALIZATION: EEG RESPONSES BY CONDITION
    # =============================================================================
    print("\n2.1 EEG Response Visualization")
    print("-"*30)
    time_samples = np.arange(eeg_8HZ.shape[2])  # Assuming this is the time dimension
    time_msec = (time_samples / sr * 1000) - 1000
    # Create comprehensive visualization
    fig = plt.figure(figsize=(20, 16))
    plt.rcParams.update({'font.size': 14})
     #min=np.mean(np.min(eeg_8HZ,1))
     #max=np.mean(np.max(eeg_8HZ,1))
    # Plot 1: Overall EEG responses (2x2 grid)
    ax1 = plt.subplot(3, 3, 1)
    avg_eeg = np.mean(eeg_8HZ[:, :, :], 0)
    plt.plot(time_msec,np.mean(avg_eeg[:, :], 0), linewidth=2, color='red')
    #plt.axvline(x=0, color='k', linestyle='-')
    plt.axhline(y=0, color='k', linestyle='--')
    plt.xlabel('Time (msec)')
    plt.ylabel('Voltage')
    plt.title('8 Hz Encoding (All Channels)')
    stim_start, stim_end = 0, 3000
    plt.axvspan(stim_start, stim_end, alpha=0.3, color='lightgrey')
    #plt.ylim(min-10,max+10)

    ax2 = plt.subplot(3, 3, 2)
    avg_eeg = np.mean(eeg_3HZ[:, :, :], 0)
    plt.plot(time_msec,np.mean(avg_eeg[:, :], 0), linewidth=2, color='blue')
    plt.axvline(x=0.1*sr, color='k', linestyle='-')
    plt.axhline(y=0, color='k', linestyle='--')
    plt.xlabel('Time (msec)')
    plt.ylabel('Voltage')
    plt.title('3 Hz Encoding (All Channels)')
    plt.axvspan(stim_start, stim_end, alpha=0.3, color='lightgrey')
    #plt.ylim(min-10,max+10)
    ax3 = plt.subplot(3, 3, 3)
    avg_eeg = np.mean(eeg_no_encoding[:, :, :], 0)
    plt.plot(time_msec,np.mean(avg_eeg[:, :], 0), linewidth=2, color='green')
    plt.axvline(x=0.1*sr, color='k', linestyle='-')
    plt.axhline(y=0, color='k', linestyle='--')
    plt.xlabel('Time (msec)')
    plt.ylabel('Voltage')
    plt.title('No Stimuli Encoding')
    plt.axvspan(stim_start, stim_end, alpha=0.3, color='lightgrey')
    global_min = np.min(np.min(eeg_8HZ,1))-10
    global_max =np.max(np.max(eeg_8HZ,1))+10
    # Plot 2: Channel-specific analysis for 8 Hz
    if Contact_stimuli_8:
        avg_eeg_other_8 = np.mean(eeg_8HZ[:, Index_other_8, :], axis=1)
        avg_eeg_stimuli_8 = np.mean(eeg_8HZ[:, Contact_stimuli_8, :], axis=1)
        avg_eeg_channel_8 = np.mean(eeg_8HZ[:, Channel_stimuli_8, :], axis=1)

        ax4 = plt.subplot(3, 3, 4)
        plt.plot(time_msec,np.mean(avg_eeg_other_8, axis=0), linewidth=2, color='blue', label='Other Channels')
        plt.axhline(y=0, color='k', linestyle='--')
        plt.xlabel('Time (msec)')
        plt.ylabel('Voltage')
        plt.title('8 Hz: Other Channels')
        plt.axvspan(stim_start, stim_end, alpha=0.3, color='lightgrey')
        plt.ylim(global_min/10,global_max/10)
        
        ax5 = plt.subplot(3, 3, 5)
        plt.plot(time_msec,np.mean(avg_eeg_stimuli_8, axis=0), linewidth=2, color='red', label='Stimulated Pairs')
        plt.axhline(y=0, color='k', linestyle='--')
        plt.xlabel('Time (msec)')
        plt.ylabel('Voltage')
        plt.title('8 Hz: Stimulated Pairs')
        plt.axvspan(stim_start, stim_end, alpha=0.3, color='lightgrey')
        plt.ylim(global_min,global_max)
        
        ax6 = plt.subplot(3, 3, 6)
        plt.plot(time_msec,np.mean(avg_eeg_channel_8, axis=0), linewidth=2, color='orange', label='Stimulated Shank')
        plt.axhline(y=0, color='k', linestyle='--')
        plt.xlabel('Time (msec)')
        plt.ylabel('Voltage')
        plt.title('8 Hz: Stimulated Shank')
        plt.axvspan(stim_start, stim_end, alpha=0.3, color='lightgrey')
        plt.ylim(global_min/10,global_max/10)
        #plt.ylim(min-10,max+10)
    # Plot 3: Channel-specific analysis for 3 Hz
    if Contact_stimuli_3:
        avg_eeg_other_3 = np.mean(eeg_3HZ[:, Index_other_3, :], axis=1)
        avg_eeg_stimuli_3 = np.mean(eeg_3HZ[:, Contact_stimuli_3, :], axis=1)
        avg_eeg_channel_3 = np.mean(eeg_3HZ[:, Channel_stimuli_3, :], axis=1)
        
        ax7 = plt.subplot(3, 3, 7)
        plt.plot(time_msec,np.mean(avg_eeg_other_3, axis=0), linewidth=2, color='blue', label='Other Channels')
        plt.axhline(y=0, color='k', linestyle='--')
        plt.xlabel('Time (msec)')
        plt.ylabel('Voltage')
        plt.title('3 Hz: Other Channels')
        plt.axvspan(stim_start, stim_end, alpha=0.3, color='lightgrey')
        plt.ylim(global_min/10,global_max/10)
        
        ax8 = plt.subplot(3, 3, 8)
        plt.plot(time_msec,np.mean(avg_eeg_stimuli_3, axis=0), linewidth=2, color='red', label='Stimulated Pairs')
        plt.axhline(y=0, color='k', linestyle='--')
        plt.xlabel('Time (msec)')
        plt.ylabel('Voltage')
        plt.title('3 Hz: Stimulated Pairs')
        plt.axvspan(stim_start, stim_end, alpha=0.3, color='lightgrey')
        plt.ylim(global_min,global_max)
        
        ax9 = plt.subplot(3, 3, 9)
        plt.plot(time_msec,np.mean(avg_eeg_channel_3, axis=0), linewidth=2, color='purple', label='Stimulated Shank')
        plt.axhline(y=0, color='k', linestyle='--')
        plt.xlabel('Time (msec)',fontsize=14)
        plt.ylabel('Voltage',fontsize=14)
        plt.title('3 Hz: Stimulated Shank',fontsize=14)
        plt.axvspan(stim_start, stim_end, alpha=0.3, color='lightgrey')
        plt.ylim(global_min/10,global_max/10)
    # =============================================================================
    # PHASE 3: CHANNEL QUALITY ASSESSMENT
    # =============================================================================
    print("\n" + "="*80)
    print("PHASE 3: CHANNEL QUALITY ASSESSMENT")
    print("-"*50)

    print("3.1 ARTIFACT DETECTION")
    print("-"*30)

    # Compute trial variability (artifact detection)
    trial_variability = np.std(eeg_no_encoding, axis=0)  # Shape: (channels, timepoints)
    timepoint_variability = np.mean(trial_variability, axis=0)  # Average across channels
    channel_variability = np.mean(trial_variability, axis=1)  # Average across timepoints

    # Z-score channel variability to identify outliers
    channel_variability_z = zscore(channel_variability)
    problematic_channels = np.where(np.abs(channel_variability_z) > 2)[0]

    print(f"Total channels analyzed: {len(channels)}")
    print(f"Channels with high variability (|z-score| > 2): {len(problematic_channels)}")

    if len(problematic_channels) > 0:
        print("\nProblematic channels:")
        for ch_idx in problematic_channels:
            if ch_idx < len(channels):
                print(f"  Channel {ch_idx}: {channels.iloc[ch_idx]['label']} (z-score: {channel_variability_z[ch_idx]:.2f})")

    # Visualization of artifact detection
    fig, axes = plt.subplots(2, 2, figsize=(16, 12))
    fig.suptitle('Channel Quality Assessment - Artifact Detection', fontsize=16)

    # Heatmap of trial variability
    ax1 = axes[0, 0]
    im = ax1.imshow(trial_variability, aspect='auto', cmap='viridis', origin='lower')
    ax1.set_xlabel('Time Points')
    ax1.set_ylabel('Channels')
    ax1.set_title('Trial Variability Heatmap')
    plt.colorbar(im, ax=ax1, label='Standard Deviation')

    # Average variability across time
    ax2 = axes[0, 1]
    ax2.plot(timepoint_variability, linewidth=2)
    ax2.set_xlabel('Time Points')
    ax2.set_ylabel('Average Standard Deviation')
    ax2.set_title('Average Trial Variability Across Time')
    ax2.grid(True, alpha=0.3)
    ax2.axvspan(stim_start, stim_end, alpha=0.3, color='lightgrey', label='Stimulus Period')
    ax2.legend()

    # Channel variability z-scores
    ax3 = axes[1, 0]
    ax3.plot(channel_variability_z, linewidth=2)
    ax3.axhline(y=2, color='r', linestyle='--', label='Z-score = ±2')
    ax3.axhline(y=-2, color='r', linestyle='--')
    ax3.set_xlabel('Channel Number')
    ax3.set_ylabel('Z-scored Standard Deviation')
    ax3.set_title('Channel Variability Z-scores')
    ax3.grid(True, alpha=0.3)
    ax3.legend()

    # Problematic channels over time
    ax4 = axes[1, 1]
    if len(problematic_channels) > 0:
        for ch in problematic_channels[:5]:  # Plot up to 5 problematic channels
            if ch < trial_variability.shape[0]:
                ax4.plot(trial_variability[ch, :], 
                        label=f'Ch {ch}: {channels.iloc[ch]["label"] if ch < len(channels) else "Unknown"}', 
                        linewidth=2)
        ax4.set_xlabel('Time Points')
        ax4.set_ylabel('Standard Deviation')
        ax4.set_title('Trial Variability for Problematic Channels')
        ax4.grid(True, alpha=0.3)
        ax4.legend()
        ax4.axvspan(stim_start, stim_end, alpha=0.3, color='lightgrey')

    plt.tight_layout()
    plt.show()
```

```python
import re

def identify_stimulated_channels(evs, indices, channels):
    """Identify stimulated channel pairs and related channels"""
    labels = []
    Stim_label = []
    all_stimuli_channel_indices = []
    all_stimuli_indices = []
    relative_index = []
    
    for ind in indices:
        stim_params = evs['stim_params'].loc[ind]
        anode_label = stim_params['anode_label']
        cathode_label = stim_params['cathode_label']
        label1 = anode_label + '-' + cathode_label
        label2 = cathode_label + '-' + anode_label
        labels.extend([label1, label2])
        Stim_label.extend([anode_label, cathode_label])
    
    labels = list(set(labels))
    Stim_label = list(set(Stim_label))
    

    # Process compound labels (anode-cathode pairs)
    for label in labels:
        # Exact match
        mask = channels['label'].str.contains(label, na=False)
        stimuli_indices = channels.loc[mask].index.tolist()
        
        # Flexible match - remove numbers from both search term and data
        cleaned_label = re.sub(r'\d+', '', label)
        mask_channel = channels['label'].str.replace(r'\d+', '', regex=True).str.contains(cleaned_label, na=False)
        stimuli_channel_indices = channels.loc[mask_channel].index.tolist()
        
        all_stimuli_indices.extend(stimuli_indices)
        all_stimuli_channel_indices.extend(stimuli_channel_indices)
    
    # Process individual electrode labels
    for label in Stim_label:
        mask = channels['label'].str.contains(label, na=False)
        individual_stimuli_indices = channels.loc[mask].index.tolist()
        relative_index.extend(individual_stimuli_indices)
    
    # Remove duplicates
    all_stimuli_indices = list(set(all_stimuli_indices))
    all_stimuli_channel_indices = list(set(all_stimuli_channel_indices))
    relative_index = list(set(relative_index))
    
    # Calculate different channel groups
    ALLother = list(range(len(channels)))
    Index_other = [x for x in ALLother if x not in all_stimuli_channel_indices]
    Contact_stimuli = all_stimuli_indices
    Channel_stimuli = [x for x in all_stimuli_channel_indices 
                      if x not in all_stimuli_indices 
                      and x not in relative_index]
    
    return Index_other, Channel_stimuli, Contact_stimuli

Index_other_8, Channel_stimuli_8, Contact_stimuli_8 = identify_stimulated_channels(evs, indices_with_8, channels)
Index_other_3, Channel_stimuli_3, Contact_stimuli_3 = identify_stimulated_channels(evs, indices_with_3, channels)
```

```python
# =============================================================================
# CONFIGURATION
# =============================================================================

# Check if required variables are defined
if 'subject' not in locals():
    raise ValueError("Variable 'subject' must be defined before running this code")
if 'exp' not in locals():
    raise ValueError("Variable 'exp' must be defined before running this code")
if 'time_window' not in locals():
    raise ValueError("Variable 'time_window' must be defined before running this code")
if 'df_select_new' not in locals():
    raise ValueError("Variable 'df_select_new' must be defined before running this code")
if 'df_select' not in locals():
    raise ValueError("Variable 'df_select' must be defined before running this code")

# Get unique sessions and iterate properly
session_list = df_select_new['session'].unique()
for i, session in enumerate(session_list):
    print("="*80)
    print(f"EEG DATA Report for subject:{subject}")
    print(f"Experiment: {exp}")
    print(f"Time window: {time_window} ms")
    print(f"Processing session {i+1} of {len(session_list)}")
    print("-"*80)
    
    # Get the current session data correctly
    df_sess = df_select_new[df_select_new['session'] == session].iloc[0]
    
    try:
        reader = cml.CMLReader(subject, exp, session=df_sess['session'],
                              localization=df_sess['localization'], 
                              montage=df_sess['montage'])
    except Exception as e:
        print(f"Error initializing CMLReader: {e}")
        continue
    
    # =============================================================================
    # PHASE 1: DATA LOGIN VERIFICATION
    # =============================================================================
    print("\nPHASE 1: DATA LOGIN VERIFICATION")
    print("-"*50)

    # Load data index and get subject data
    if df_select.empty:
        raise ValueError(f"No data found for subject {subject} in experiment {exp}")

    # Initialize reader and load events
    try:
        evs = reader.load('task_events')
    except Exception as e:
        print(f"Error loading task events: {e}")
        continue

    print("1.1 EVENT VERIFICATION")
    print("-"*30)

    # Define expected event counts
    expected_counts = {
        'START': 1,
        'store mappings': 1,
        'SESS_END': 1,
        'TRIAL_START': 5,
        'TRIAL_END': 5,
        'WORD': 60,
        'STIM': '89 or 91',
        'pointing begins': 65,
        'pointing finished': 65,
        'REC_START': 5,
        'REC_STOP': 5,
        'CUED_REC_CUE': 60,
        'CUED_REC_STOP': 60
    }

    # Verify event counts
    event_verification = {}
    for event_type, expected in expected_counts.items():
        actual = len(evs[evs['type'] == event_type])
        event_verification[event_type] = {
            'actual': actual,
            'expected': expected,
            'status': 'PASS' if (isinstance(expected, int) and actual == expected) or 
                               (isinstance(expected, str) and actual in [89, 91]) else 'FAIL'
        }
        print(f"{event_type}: {actual} / {expected} - {event_verification[event_type]['status']}")

    print("\n1.2 STIMULUS DELIVERY VALIDATION")
    print("-"*30)

    # Extract stimulus parameters with error handling
    pulse_freq_data = []
    word_data_type = []
    
    for i in range(len(evs['stim_params'])):
        try:
            freq_data = evs['stim_params'].iloc[i]['burst_freq']
            word_data = evs['type'].iloc[i]
            pulse_freq_data.append(freq_data)
            word_data_type.append(word_data)
        except (KeyError, IndexError) as e:
            print(f"Warning: Error extracting stimulus parameters at index {i}: {e}")
            pulse_freq_data.append(-1)  # Default value
            word_data_type.append('UNKNOWN')

    # Find stimulus indices with bounds checking
    pulse_freq_array = np.array(pulse_freq_data)
    indices_with_8 = np.where(pulse_freq_array == 8)[0].tolist()
    indices_with_3 = np.where(pulse_freq_array == 3)[0].tolist()
    
    # Fixed indices_with_no calculation
    if len(pulse_freq_array) > 1:
        no_stim_indices = np.where(pulse_freq_array[:-1] == -1)[0]
        indices_with_no = []
        for idx in no_stim_indices:
            if idx + 1 < len(pulse_freq_array):
                indices_with_no.append(idx + 1)
    else:
        indices_with_no = []

    print(f"8 Hz stimuli events: {len(indices_with_8)}")
    print(f"3 Hz stimuli events: {len(indices_with_3)}")
    print(f"No stimuli events: {len(indices_with_no)}")

    # Add phase information to events
    def add_phase(evs):
        new_evs = pd.DataFrame([])
        for ind, list_evs in evs.groupby('trial'):  # 'trial' is the equivalent of a delivery day
            list_evs = list_evs.copy()  # Avoid SettingWithCopyWarning
            if list_evs['trial'].unique()[0] != -999:
                list_evs['phase'] = 'nan'
                
                if not list_evs[list_evs['type'] == 'TRIAL_START'].index.empty:
                    # find indexes of encoding start and end
                    trial_start_events = list_evs[list_evs['type'] == 'TRIAL_START']
                    trial_end_events = list_evs[list_evs['type'] == 'TRIAL_END']
                    
                    if not trial_start_events.empty and not trial_end_events.empty:
                        first_pointing_index = trial_start_events.index.min()
                        navigation_end_index = trial_end_events.index.min() - 1
                        list_evs.loc[first_pointing_index:navigation_end_index, 'phase'] = 'encoding'
                    
                    # find indexes of retrieval start and end
                    rec_start_events = list_evs[list_evs['type'] == 'REC_START']
                    rec_stop_events = list_evs[list_evs['type'] == 'REC_STOP']
                    
                    if not rec_start_events.empty and not rec_stop_events.empty:
                        rec_start_index = rec_start_events.index[0]
                        rec_end_index = rec_stop_events.index[0]
                        list_evs.loc[rec_start_index:rec_end_index, 'phase'] = 'retrieval'

                    if len(list_evs[list_evs.type == 'FSR_START']) > 0:  # relevant only for the final trial
                        # find indexes of Final FR for stores start and end
                        fsr_start_events = list_evs[list_evs['type'] == 'FSR_START']
                        fsr_stop_events = list_evs[list_evs['type'] == 'FSR_STOP']
                        
                        if not fsr_start_events.empty and not fsr_stop_events.empty:
                            FSR_start_index = fsr_start_events.index[0]
                            FSR_end_index = fsr_stop_events.index[0]
                            list_evs.loc[FSR_start_index:FSR_end_index, 'phase'] = 'FSR'

                        # find indexes of Final FR for items start and end
                        ffr_start_events = list_evs[list_evs['type'] == 'FFR_START']
                        ffr_stop_events = list_evs[list_evs['type'] == 'FFR_STOP']
                        
                        if not ffr_start_events.empty and not ffr_stop_events.empty:
                            FFR_start_index = ffr_start_events.index[0]
                            FFR_end_index = ffr_stop_events.index[0]
                            list_evs.loc[FFR_start_index:FFR_end_index, 'phase'] = 'FFR'

            elif list_evs['trial'].unique()[0] == -999:
                list_evs['phase'] = 'nan'

            new_evs = pd.concat([new_evs, list_evs])
        return new_evs
    
    evs = add_phase(evs)
    
    # Analyze stimuli by phase with error handling
    try:
        result_8HZ_encoding = evs[(evs.index.isin(indices_with_8)) & (evs['phase'] == 'encoding')]
        result_3HZ_encoding = evs[(evs.index.isin(indices_with_3)) & (evs['phase'] == 'encoding')] 
        result_none_encoding = evs[(evs.index.isin(indices_with_no)) & (evs['phase'] == 'encoding') & (evs['type'] == 'WORD')]
        result_8HZ_retrieval = evs[(evs.index.isin(indices_with_8)) & (evs['phase'] == 'retrieval')]
        result_3HZ_retrieval = evs[(evs.index.isin(indices_with_3)) & (evs['phase'] == 'retrieval')]
        
        print(f"\nStimuli distribution by phase:")
        print(f"Encoding - 3 Hz: {len(result_3HZ_encoding)}, 8 Hz: {len(result_8HZ_encoding)}, None: {len(result_none_encoding)}")
        print(f"Retrieval - 3 Hz: {len(result_3HZ_retrieval)}, 8 Hz: {len(result_8HZ_retrieval)}")
    except Exception as e:
        print(f"Error analyzing stimuli by phase: {e}")
        continue

    # =============================================================================
    # PHASE 2: FREQUENCY ANALYSIS (3-8 Hz Peak Detection)
    # =============================================================================
    print("\n" + "="*80)
    print("PHASE 2: FREQUENCY ANALYSIS (3-8 Hz Peak Detection)")
    print("-"*50)

    # Load EEG data - reader is already initialized
    try:
        evs_reload = reader.load("events")  # Use different variable name to avoid overwriting
        channels = reader.load("pairs")
    except Exception as e:
        print(f"Error loading events or pairs: {e}")
        continue

    # Validate that we have events to process
    if result_8HZ_encoding.empty and result_3HZ_encoding.empty and result_none_encoding.empty:
        print("Warning: No events found for any condition. Skipping EEG loading.")
        continue

    # Load EEG for different conditions with error handling
    print("Loading EEG data...")
    eeg_data = {}
    
    try:
        if not result_8HZ_encoding.empty:
            eeg_container_8HZ_encoding = reader.load_eeg(result_8HZ_encoding, -1000, time_window, scheme=channels)
            eeg_data['8HZ_encoding'] = eeg_container_8HZ_encoding.data
            sr = eeg_container_8HZ_encoding.samplerate
        else:
            eeg_data['8HZ_encoding'] = None
            
        if not result_3HZ_encoding.empty:
            eeg_container_3HZ_encoding = reader.load_eeg(result_3HZ_encoding, -1000, time_window, scheme=channels)
            eeg_data['3HZ_encoding'] = eeg_container_3HZ_encoding.data
            if 'sr' not in locals():
                sr = eeg_container_3HZ_encoding.samplerate
        else:
            eeg_data['3HZ_encoding'] = None
            
        if not result_8HZ_retrieval.empty:
            eeg_container_8HZ_retrieval = reader.load_eeg(result_8HZ_retrieval, -1000, time_window, scheme=channels)
            eeg_data['8HZ_retrieval'] = eeg_container_8HZ_retrieval.data
        else:
            eeg_data['8HZ_retrieval'] = None
            
        if not result_3HZ_retrieval.empty:
            eeg_container_3HZ_retrieval = reader.load_eeg(result_3HZ_retrieval, -1000, time_window, scheme=channels)
            eeg_data['3HZ_retrieval'] = eeg_container_3HZ_retrieval.data
        else:
            eeg_data['3HZ_retrieval'] = None
            
        if not result_none_encoding.empty:
            eeg_container_no_encoding = reader.load_eeg(result_none_encoding, -1000, time_window, scheme=channels)
            eeg_data['no_encoding'] = eeg_container_no_encoding.data
        else:
            eeg_data['no_encoding'] = None
            
    except Exception as e:
        print(f"Error loading EEG data: {e}")
        continue

    # Extract data arrays for backward compatibility
    eeg_8HZ = eeg_data['8HZ_encoding']
    eeg_3HZ = eeg_data['3HZ_encoding']
    eeg_8HZ_retrieval = eeg_data['8HZ_retrieval']
    eeg_3HZ_retrieval = eeg_data['3HZ_retrieval']
    eeg_no_encoding = eeg_data['no_encoding']

    print(f"Sample rate: {sr} Hz")
    print(f"EEG data shapes:")
    if eeg_8HZ is not None:
        print(f"  8 Hz encoding: {eeg_8HZ.shape}")
    if eeg_3HZ is not None:
        print(f"  3 Hz encoding: {eeg_3HZ.shape}")
    if eeg_no_encoding is not None:
        print(f"  No stimuli encoding: {eeg_no_encoding.shape}")

    # Identify stimulated channels with improved error handling
    def identify_stimulated_channels(evs, indices, channels):
        """Identify stimulated channel pairs and related channels"""
        if not indices:  # Check if indices list is empty
            print("Warning: No stimulus indices provided")
            return [], [], []
            
        labels = []
        Stim_label = []
        
        for ind in indices:
            try:
                stim_params = evs['stim_params'].iloc[ind]
                anode_label = stim_params['anode_label']
                cathode_label = stim_params['cathode_label']
                label1 = anode_label + '-' + cathode_label
                label2 = cathode_label + '-' + anode_label
                labels.extend([label1, label2])
                Stim_label.extend([anode_label, cathode_label])
            except (KeyError, IndexError) as e:
                print(f"Warning: Error extracting stim_params at index {ind}: {e}")
                continue
                
        labels = list(set(labels))
        Stim_label = list(set(Stim_label))  # Remove duplicates

        all_stimuli_channel_indices = []
        all_stimuli_indices = []
        relative_index = []
        
        for label in labels:
            # Exact match
            mask = channels['label'].str.contains(label, na=False, regex=False)
            stimuli_indices = channels.loc[mask].index.tolist()
          
            # Flexible match - remove numbers from both search term and data
            cleaned_label = re.sub(r'\d+', '', label)
            mask_channel = channels['label'].str.replace(r'\d+', '', regex=True).str.contains(cleaned_label, na=False, regex=False)
            stimuli_channel_indices = channels.loc[mask_channel].index.tolist()

            all_stimuli_indices.extend(stimuli_indices)
            all_stimuli_channel_indices.extend(stimuli_channel_indices)
            
        for label in Stim_label:
            # Exact match
            mask = channels['label'].str.contains(label, na=False, regex=False)
            stimuli_indices = channels.loc[mask].index.tolist()
            relative_index.extend(stimuli_indices)  # Fixed: was extending with itself
        
        all_stimuli_indices = list(set(all_stimuli_indices))
        all_stimuli_channel_indices = list(set(all_stimuli_channel_indices))
        relative_index = list(set(relative_index))
        
        # Validate indices are within bounds
        max_channels = len(channels)
        all_stimuli_indices = [idx for idx in all_stimuli_indices if 0 <= idx < max_channels]
        all_stimuli_channel_indices = [idx for idx in all_stimuli_channel_indices if 0 <= idx < max_channels]
        relative_index = [idx for idx in relative_index if 0 <= idx < max_channels]
        
        ALLother = list(range(0, len(channels)))  # Fixed: start from 0
        Index_other = [x for x in ALLother if x not in all_stimuli_channel_indices]
        Contact_stimuli = all_stimuli_indices
        Channel_stimuli = [x for x in all_stimuli_channel_indices 
                          if x not in all_stimuli_indices 
                          and x not in relative_index]
        
        return Index_other, Channel_stimuli, Contact_stimuli

    # Identify channels for both frequencies
    Index_other_8, Channel_stimuli_8, Contact_stimuli_8 = identify_stimulated_channels(evs, indices_with_8, channels)
    Index_other_3, Channel_stimuli_3, Contact_stimuli_3 = identify_stimulated_channels(evs, indices_with_3, channels)

    print(f"\n8 Hz stimulated contacts: {Contact_stimuli_8}")
    print(f"3 Hz stimulated contacts: {Contact_stimuli_3}")

    # =============================================================================
    # VISUALIZATION: EEG RESPONSES BY CONDITION
    # =============================================================================
    print("\n2.1 EEG Response Visualization")
    print("-"*30)

    # Create comprehensive visualization only if we have data
    if any(data is not None for data in [eeg_8HZ, eeg_3HZ, eeg_no_encoding]):
        fig = plt.figure(figsize=(20, 16))

        # Plot 1: Overall EEG responses (3x3 grid)
        if eeg_8HZ is not None:
            ax1 = plt.subplot(3, 3, 1)
            avg_eeg = np.mean(eeg_8HZ[:, :, :], 0)
            plt.plot(np.mean(avg_eeg[:, :], 0), linewidth=2, color='red')
            plt.axvline(x=0.1*sr, color='k', linestyle='-')
            plt.axhline(y=0, color='k', linestyle='--')
            plt.xlabel('Time (samples)')
            plt.ylabel('Voltage')
            plt.title('8 Hz Encoding (All Channels)')
            stim_start, stim_end = 1000, 4000
            plt.axvspan(stim_start, stim_end, alpha=0.3, color='lightgrey')

        if eeg_3HZ is not None:
            ax2 = plt.subplot(3, 3, 2)
            avg_eeg = np.mean(eeg_3HZ[:, :, :], 0)
            plt.plot(np.mean(avg_eeg[:, :], 0), linewidth=2, color='blue')
            plt.axvline(x=0.1*sr, color='k', linestyle='-')
            plt.axhline(y=0, color='k', linestyle='--')
            plt.xlabel('Time (samples)')
            plt.ylabel('Voltage')
            plt.title('3 Hz Encoding (All Channels)')
            plt.axvspan(stim_start, stim_end, alpha=0.3, color='lightgrey')

        if eeg_no_encoding is not None:
            ax3 = plt.subplot(3, 3, 3)
            avg_eeg = np.mean(eeg_no_encoding[:, :, :], 0)
            plt.plot(np.mean(avg_eeg[:, :], 0), linewidth=2, color='green')
            plt.axvline(x=0.1*sr, color='k', linestyle='-')
            plt.axhline(y=0, color='k', linestyle='--')
            plt.xlabel('Time (samples)')
            plt.ylabel('Voltage')
            plt.title('No Stimuli Encoding')
            plt.axvspan(stim_start, stim_end, alpha=0.3, color='lightgrey')
      
        # Plot 2: Channel-specific analysis for 8 Hz
        if eeg_8HZ is not None and Contact_stimuli_8:
            # Validate channel indices before indexing
            valid_other_8 = [idx for idx in Index_other_8 if idx < eeg_8HZ.shape[1]]
            valid_contact_8 = [idx for idx in Contact_stimuli_8 if idx < eeg_8HZ.shape[1]]
            valid_channel_8 = [idx for idx in Channel_stimuli_8 if idx < eeg_8HZ.shape[1]]
            
            if valid_other_8:
                avg_eeg_other_8 = np.mean(eeg_8HZ[:, valid_other_8, :], axis=1)
                ax4 = plt.subplot(3, 3, 4)
                plt.plot(np.mean(avg_eeg_other_8, axis=0), linewidth=2, color='blue', label='Other Channels')
                plt.axhline(y=0, color='k', linestyle='--')
                plt.xlabel('Time (samples)')
                plt.ylabel('Voltage')
                plt.title('8 Hz: Other Channels')
                plt.axvspan(stim_start, stim_end, alpha=0.3, color='lightgrey')

            if valid_contact_8:
                avg_eeg_stimuli_8 = np.mean(eeg_8HZ[:, valid_contact_8, :], axis=1)
                ax5 = plt.subplot(3, 3, 5)
                plt.plot(np.mean(avg_eeg_stimuli_8, axis=0), linewidth=2, color='red', label='Stimulated Pairs')
                plt.axhline(y=0, color='k', linestyle='--')
                plt.xlabel('Time (samples)')
                plt.ylabel('Voltage')
                plt.title('8 Hz: Stimulated Pairs')
                plt.axvspan(stim_start, stim_end, alpha=0.3, color='lightgrey')

            if valid_channel_8:
                avg_eeg_channel_8 = np.mean(eeg_8HZ[:, valid_channel_8, :], axis=1)
                ax6 = plt.subplot(3, 3, 6)
                plt.plot(np.mean(avg_eeg_channel_8, axis=0), linewidth=2, color='orange', label='Stimulated Shank')
                plt.axhline(y=0, color='k', linestyle='--')
                plt.xlabel('Time (samples)')
                plt.ylabel('Voltage')
                plt.title('8 Hz: Stimulated Shank')
                plt.axvspan(stim_start, stim_end, alpha=0.3, color='lightgrey')

        # Plot 3: Channel-specific analysis for 3 Hz
        if eeg_3HZ is not None and Contact_stimuli_3:
            # Validate channel indices before indexing
            valid_other_3 = [idx for idx in Index_other_3 if idx < eeg_3HZ.shape[1]]
            valid_contact_3 = [idx for idx in Contact_stimuli_3 if idx < eeg_3HZ.shape[1]]
            valid_channel_3 = [idx for idx in Channel_stimuli_3 if idx < eeg_3HZ.shape[1]]
            
            if valid_other_3:
                avg_eeg_other_3 = np.mean(eeg_3HZ[:, valid_other_3, :], axis=1)
                ax7 = plt.subplot(3, 3, 7)
                plt.plot(np.mean(avg_eeg_other_3, axis=0), linewidth=2, color='blue', label='Other Channels')
                plt.axhline(y=0, color='k', linestyle='--')
                plt.xlabel('Time (samples)')
                plt.ylabel('Voltage')
                plt.title('3 Hz: Other Channels')
                plt.axvspan(stim_start, stim_end, alpha=0.3, color='lightgrey')

            if valid_contact_3:
                avg_eeg_stimuli_3 = np.mean(eeg_3HZ[:, valid_contact_3, :], axis=1)
                ax8 = plt.subplot(3, 3, 8)
                plt.plot(np.mean(avg_eeg_stimuli_3, axis=0), linewidth=2, color='red', label='Stimulated Pairs')
                plt.axhline(y=0, color='k', linestyle='--')
                plt.xlabel('Time (samples)')
                plt.ylabel('Voltage')
                plt.title('3 Hz: Stimulated Pairs')
                plt.axvspan(stim_start, stim_end, alpha=0.3, color='lightgrey')

            if valid_channel_3:
                avg_eeg_channel_3 = np.mean(eeg_3HZ[:, valid_channel_3, :], axis=1)
                ax9 = plt.subplot(3, 3, 9)
                plt.plot(np.mean(avg_eeg_channel_3, axis=0), linewidth=2, color='purple', label='Stimulated Shank')
                plt.axhline(y=0, color='k', linestyle='--')
                plt.xlabel('Time (samples)')
                plt.ylabel('Voltage')
                plt.title('3 Hz: Stimulated Shank')
                plt.axvspan(stim_start, stim_end, alpha=0.3, color='lightgrey')

        plt.tight_layout()
        plt.show()

    # =============================================================================
    # PHASE 3: CHANNEL QUALITY ASSESSMENT
    # =============================================================================
    print("\n" + "="*80)
    print("PHASE 3: CHANNEL QUALITY ASSESSMENT")
    print("-"*50)

    print("3.1 ARTIFACT DETECTION")
    print("-"*30)

    # Only perform quality assessment if we have no-stimuli data
    if eeg_no_encoding is not None:
        # Compute trial variability (artifact detection)
        trial_variability = np.std(eeg_no_encoding, axis=0)  # Shape: (channels, timepoints)
        timepoint_variability = np.mean(trial_variability, axis=0)  # Average across channels
        channel_variability = np.mean(trial_variability, axis=1)  # Average across timepoints

        # Z-score channel variability to identify outliers
        if len(channel_variability) > 1:  # Need at least 2 values for zscore
            channel_variability_z = zscore(channel_variability)
            problematic_channels = np.where(np.abs(channel_variability_z) > 2)[0]
        else:
            channel_variability_z = np.array([0])
            problematic_channels = np.array([])

        print(f"Total channels analyzed: {len(channels)}")
        print(f"Channels with high variability (|z-score| > 2): {len(problematic_channels)}")

        if len(problematic_channels) > 0:
            print("\nProblematic channels:")
            for ch_idx in problematic_channels:
                if ch_idx < len(channels):
                    print(f"  Channel {ch_idx}: {channels.iloc[ch_idx]['label']} (z-score: {channel_variability_z[ch_idx]:.2f})")

        # Visualization of artifact detection
        fig, axes = plt.subplots(2, 2, figsize=(16, 12))
        fig.suptitle('Channel Quality Assessment - Artifact Detection', fontsize=16)

        # Heatmap of trial variability
        ax1 = axes[0, 0]
        im = ax1.imshow(trial_variability, aspect='auto', cmap='viridis', origin='lower')
        ax1.set_xlabel('Time Points')
        ax1.set_ylabel('Channels')
        ax1.set_title('Trial Variability Heatmap')
        plt.colorbar(im, ax=ax1, label='Standard Deviation')

        # Average variability across time
        ax2 = axes[0, 1]
        ax2.plot(timepoint_variability, linewidth=2)
        ax2.set_xlabel('Time Points')
        ax2.set_ylabel('Average Standard Deviation')
        ax2.set_title('Average Trial Variability Across Time')
        ax2.grid(True, alpha=0.3)
        stim_start, stim_end = 1000, 4000
        ax2.axvspan(stim_start, stim_end, alpha=0.3, color='lightgrey', label='Stimulus Period')
        ax2.legend()

        # Channel variability z-scores
        ax3 = axes[1, 0]
        ax3.plot(channel_variability_z, linewidth=2)
        ax3.axhline(y=2, color='r', linestyle='--', label='Z-score = ±2')
        ax3.axhline(y=-2, color='r', linestyle='--')
        ax3.set_xlabel('Channel Number')
        ax3.set_ylabel('Z-scored Standard Deviation')
        ax3.set_title('Channel Variability Z-scores')
        ax3.grid(True, alpha=0.3)
        ax3.legend()

        # Problematic channels over time
      
```

```python
def identify_stimulated_channels(evs, indices, channels):
    """Identify stimulated channel pairs and related channels"""
    stimulated_pairs = []
    stimulated_contacts = []
    
    # Extract stimulated electrode labels
    for ind in indices:
        stim_params = evs['stim_params'].loc[ind]
        anode_label = stim_params['anode_label']
        cathode_label = stim_params['cathode_label']
        
        # Store the exact stimulated pair
        pair_label1 = anode_label + '-' + cathode_label
        pair_label2 = cathode_label + '-' + anode_label
        stimulated_pairs.extend([pair_label1, pair_label2])
        
        # Store individual stimulated contacts
        stimulated_contacts.extend([anode_label, cathode_label])
    
    # Remove duplicates
    stimulated_pairs = list(set(stimulated_pairs))
    stimulated_contacts = list(set(stimulated_contacts))
    
    # Find indices for each category
    contact_stimuli_indices = []  # Exact stimulated pairs
    channel_stimuli_indices = []  # Channels on same shank/related to stimulated contacts
    relative_indices = []         # Individual stimulated contacts
    
    # 1. Find exact stimulated pairs (Contact_stimuli)
    for pair_label in stimulated_pairs:
        mask = channels['label'].str.contains(pair_label, na=False, regex=False)
        contact_indices = channels.loc[mask].index.tolist()
        contact_stimuli_indices.extend(contact_indices)
    
    # 2. Find individual stimulated contacts (relative_index)
    for contact_label in stimulated_contacts:
        mask = channels['label'].str.contains(contact_label, na=False, regex=False)
        contact_indices = channels.loc[mask].index.tolist()
        relative_indices.extend(contact_indices)
    
    # 3. Find channels on same shank (Channel_stimuli)
    # Remove numbers from contact labels to find related channels on same shank
    for contact_label in stimulated_contacts:
        cleaned_label = re.sub(r'\d+', '', contact_label)  # Remove numbers
        if cleaned_label:  # Make sure we have a valid label after cleaning
            mask = channels['label'].str.replace(r'\d+', '', regex=True).str.contains(
                cleaned_label, na=False, regex=False)
            shank_indices = channels.loc[mask].index.tolist()
            channel_stimuli_indices.extend(shank_indices)
    
    # Remove duplicates and ensure proper categorization
    contact_stimuli_indices = list(set(contact_stimuli_indices))
    relative_indices = list(set(relative_indices))
    channel_stimuli_indices = list(set(channel_stimuli_indices))
    
    # Channel_stimuli should exclude exact pairs and individual contacts
    channel_stimuli_indices = [x for x in channel_stimuli_indices 
                              if x not in contact_stimuli_indices 
                              and x not in relative_indices]
    
    # Index_other includes all other channels
    all_channels = list(range(len(channels)))
    all_stimulated = list(set(contact_stimuli_indices + channel_stimuli_indices + relative_indices))
    index_other = [x for x in all_channels if x not in all_stimulated]
    
    return index_other, channel_stimuli_indices, contact_stimuli_indices
```
