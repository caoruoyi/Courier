```python
import numpy as np
import pandas as pd

def extract_xyz_coordinates(tal_struct):
    """
    Extract X, Y, Z coordinates from tal_struct['al'] record array.
    
    Returns:
    - coords_list: List of [x, y, z] coordinates
    - coords_array: NumPy array of coordinates
    - coords_df: Pandas DataFrame with columns ['x', 'y', 'z', 'region']
    """
    
    coords_list = []
    regions_list = []
    
    # Iterate through each record in tal_struct['al']
    for record in tal_struct['atlases']:
        # Each record contains multiple tuples
        for item in record:
            if isinstance(item, tuple) and len(item) >= 4:
                try:
                    # Extract region name (first element)
                    region = item[0] if item[0] != 'None' else None
                    
                    # Extract coordinates (elements 1, 2, 3)
                    x, y, z = float(item[1]), float(item[2]), float(item[3])
                    
                    # Skip NaN values
                    if not (np.isnan(x) or np.isnan(y) or np.isnan(z)):
                        coords_list.append([x, y, z])
                        regions_list.append(region)
                        
                except (ValueError, TypeError):
                    # Skip items that can't be converted to float
                    continue
    
    # Convert to numpy array
    coords_array = np.array(coords_list)
    
    # Create DataFrame
    coords_df = pd.DataFrame(coords_array, columns=['x', 'y', 'z'])
    coords_df['region'] = regions_list
    
    return coords_list, coords_array, coords_df

# Usage examples:
coords_list, coords_array, coords_df = extract_xyz_coordinates(tal_struct)

# Print summary
print(f"Total coordinates extracted: {len(coords_list)}")
print(f"Coordinate ranges:")
print(f"  X: {coords_array[:, 0].min():.2f} to {coords_array[:, 0].max():.2f}")
print(f"  Y: {coords_array[:, 1].min():.2f} to {coords_array[:, 1].max():.2f}")
print(f"  Z: {coords_array[:, 2].min():.2f} to {coords_array[:, 2].max():.2f}")

# Show first few coordinates
print("\nFirst 10 coordinates:")
print(coords_df.head(10))

# Access coordinates in different formats:
# As list: coords_list
# As numpy array: coords_array  
# As pandas DataFrame: coords_df

# Examples of accessing data:
print(f"\nFirst coordinate as list: {coords_list[0]}")
print(f"All X coordinates: {coords_array[:, 0]}")
print(f"Coordinates with region info:\n{coords_df.head()}")
```
